## Numpy version

```python
# N is batch size, D is input and output dimension, H is hidden dimension
N, D_in, H, D_out = 64, 1000, 100, 10

# Create random input and output data
x = np.random.randn(N, D_in)
y = np.random.randn(N, D_out)

# Randomly initialize weights
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)

lr = 1e-6
epoch = 500
for t in range(epoch):
    # Forward pass: compute predicted y
    h = x.dot(w1)
    h_relu = np.maximum(h, 0)
    y_pred = h_relu.dot(w2)
    
    # Compute and print loss
    loss = np.square(y_pred - y).sum()
    print(t, loss)
    
    # Backprop to compute gradients of w1 and w2 with respect to loss
    # grad_y_pred is gradient of y_pred
    grad_y_pred = 2.0 * (y_pred - y)
    # grad_w2 is gradient of w2
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h < 0] = 0
    grad_w1 = x.T.dot(grad_h)
    
    # Update weights
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
```

## Pytorch version like numpy

```python
dtype = torch.FloatTensor
N, D_in, H, D_out = 64, 1000, 100, 10
x = torch.randn(N, D_in).type(dtype)
y = torch.randn(N, D_out).type(dtype)
w1 = torch.randn(D_in, H).type(dtype)
w2 = torch.randn(H, D_out).type(dtype)
lr = 1e-6
epoch = 500
for t in range(epoch):
        #Forward pass
        h = x.mm(w1) # mm is matmul
        h_relu = h.clamp(min=0) # clamp can restrict min and max values.
        y_pred = h_relu.mm(w2)
        
        # Compute and print loss
        loss = (y_pred - y).pow(2).sum()
        print(t, loss)
        
        # Backprop
        grad_y_pred = 2.0 * (y_pred - y)
        grad_w2 = h_relu.t().mm(grad_y_pred)
        grad_h_relu = grad_y_pred.mm(w2.t())
        grad_h = grad_h_relu.clone()
        grad_h[h < 0] = 0
        grad_w1 = x.t().mm(grad_h)
        
        # Update weights using gradient descent
        w1 -= lr * grad_w1
        w2 -= lr * grad_w2
```

## pytorch using autograd version

```python
dtype = torch.FloatTensor
N, D_in, H, D_out = 64, 1000, 100, 10
x = Variable(torch.randn(N, D_in).type(dtype), requires_grad=False)
y = Variable(torch.randn(N, D_out).type(dtype), requires_grad=False)
# requires_grad=False indicates that we do not need to compute gradients with respect to these Variables during the backward pass.

w1 = Variable(torch.randn(D_in, H).type(dtype), requires_grad=True)
w2 = Variable(torch.randn(H, D_out).type(dtype), requires_grad=True)

lr = 1e-6
epoch = 500
for t in range(500):
    # Forward pass
    # We do not need to make intermediate values because of not implementing the backward pass by hand.
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    
    loss = (y_pred - y).pow(2).sum()
    print(t, loss.data[0])
    
    # Use autograd to compute the backward pass.
    # This call will compute the gradient of loss with repect to all Variables with requires_grad=True.
    # After calling backward, you can use w1.grad and w2.grad holding the gradient each variables.
    loss.backward()
    
    # Update weights using gardient descent
    # w1.grad is Variable, but w1.grad.data is Tensor.
    w1.data -= lr * w1.grad.data
    w2.data -= lr * w2.grad.data
    
    # Don't forget to manually make the gradients zero after updating weights.
    w1.grad.data.zero_()
    w2.grad.data.zero_()
```

## pytorch using nn module version

```python
N, D_in, H, D_out = 64, 1000, 100, 10
x = Variable(torch.randn(N, D_in))
y = Variable(torch.randn(N, D_out), requires_grad=False)

# Use the nn package, and computes output from input using a linear function.
# nn.Sequential contains other modules.
# Each linear Module holds internal Variables for its weight and bias.
model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
)

# The nn package contains definitions of popular loss functions
loss_fn = torch.nn.MSELoss(size_average=False)

lr = 1e-4
epoch = 500
for t in range(epoch):
    # Forwarc pass: copute predicted y
    # Module objects override the __call__ operator, so you can call them like functions.
    # When doing so you pass a Variable of input data to the Module
    # and it produces a Variabel of output data.
    y_pred = model(x)
    
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    # Make the gradients zero before running the backward pass.
    model.zero_grad()
    
    # Backward pass: compute gradient of the loss
    # with respect to all the lernable parameters in the model.
    loss.backward()
    
    # Update the weights using gradient descent.
    for param in model.parameters():
        param.data -= lr * param.grad.data
```

## pytorch using optim version

```python
N, D_in, H, D_out = 64, 1000, 100, 10
x = Variable(torch.randn(N, D_in))
y = Variable(torch.randn(N, D_out), requires_grad=False)

model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
)

loss_fn = torch.nn.MSELoss(size_average=False)
lr = 1e-4
epoch = 500
# Use the optim package to define an Optimizer.
# The first augument ot the Adam constructor tells the optimizer
# which Variables it should update
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
for t in range(epoch):
    # Forward pass
    y_pred = model(x)
    
    # Compute loss
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    # Before the backward pass, use the optimizer object to zero all of the gradients
    # for the variables it will update.
    optimizer.zero_grad()
    
    # Backward pass
    loss.backward()
    # Calling the step function on an Optimizer makes an update to its parameters
    optimizer.step()
```

## pytorch using custom nn modules version

```python
class TwoLayerNet(torch.nn.Module):
    # In the constractor, which is the function to make new instance,
    # we instantiate two nn.Linear modules and assign them as member variables.
    def __init__(self, D_in, H, D_out): # This is constractor.
        super(TwoLayerNet, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2 = torch.nn.Linear(H, D_out)
        
    # In the forward function 
    # accept a Variable of input data
    # return a Variable of output data
    # We can use Modules defined in the constructor as well as arbitrary operators on Variables.
    def forward(self, x):
        h_relu = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(h_relu)
        return y_pred
    
N, D_in, H, D_out = 64, 1000, 100, 10
x = Variable(torch.randn(N, D_in))
y = Variable(torch.randn(N, D_out), requires_grad=False)

# Constract our model by instantiating the class defined above
model = TwoLayerNet(D_in, H, D_out)
# Construct our loss function. (criterion is base.)
criterion = torch.nn.MSELoss(size_average=False)
# COnstract our optimizer .
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4)
for t in range(500):
    # Forward pass
    y_pred = model(x)

    # Compute loss
    loss = criterion(y_pred, y)
    print(t, loss.data[0])

    # Make the gradients zero
    optimizer.zero_grad()
    
    # Backward pass
    loss.backward()
    
    # Update the weights
    optimizer.step()
```
